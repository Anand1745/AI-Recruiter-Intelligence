```python
import json
import pandas as pd

rows = []

with open("../data/candidates.jsonl", "r", encoding="utf-8") as f:

    for line in f:

        c = json.loads(line)

        s = c.get("redrob_signals", {})

        rows.append({

            "response_rate":
                s.get("recruiter_response_rate"),

            "github":
                s.get("github_activity_score"),

            "profile":
                s.get("profile_completeness_score"),

            "connections":
                s.get("connection_count"),

            "interview":
                s.get("interview_completion_rate"),

            "offer":
                s.get("offer_acceptance_rate"),

            "open_to_work":
                s.get("open_to_work_flag"),

            "last_active":
                s.get("last_active_date")
        })

df = pd.DataFrame(rows)

df.describe(include="all")
```
