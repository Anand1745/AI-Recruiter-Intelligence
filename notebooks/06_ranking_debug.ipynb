```python
import json

found = False

with open("../data/candidates.jsonl", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        c = json.loads(line)

        name = c["profile"]["anonymized_name"]

        if "Ved" in name or "Bhatia" in name:
            print("Found!")
            print("Row:", i + 1)
            print(name)
            found = True
            break

if not found:
    print("Ved Bhatia not found.")
```

```python
import json

with open("../data/candidates.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        c = json.loads(line)

        if c["candidate_id"] == "CAND_0000077":
            ved = c
            break

print(ved)
```

```python
profile = pipeline.processor.process_profile(ved)
skills = pipeline.processor.process_skills(ved)

print(profile)
print()
print(skills)
```

```python
parsed = pipeline.parser.parse(jd)

score = pipeline.matcher.similarity(
    jd,
    profile["headline"] + " " +
    profile["summary"] + " " +
    skills["skills_text"]
)

ranking = pipeline.ranker.rank(
    score,
    profile,
    skills,
    parsed
)

ranking
```
